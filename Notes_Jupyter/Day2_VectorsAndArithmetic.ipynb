## Definition

A **vector** is a quantity with both **magnitude** and **direction**. Vectors are often represented as arrays of numbers:

$$\vec{v} = \left[\begin{array}{r} v_1 \\ v_2 \\ \vdots \\ v_n \end{array}\right]  \text{(column vector)} \qquad \vec{u} = \left[\begin{array}{r} u_1 & u_2 & \cdots & u_n \end{array}\right] \quad \text{(row vector)}$$

> **Example:** Consider the vectors $\vec{v} = \begin{bmatrix} 1\\ -8\\ 3/2\end{bmatrix}$ and $\vec{u} = \begin{bmatrix} -11 & 17 & 0 & \pi\end{bmatrix}$

## Transposing Vectors

We can turn a *column vector* into a *row vector* (and vice-versa) by taking its **transpose**. We denote the transpose operation via the superscript $^T$. Consider the vector $\vec{v}$ and its transpose below.

$$\vec{v} = \begin{bmatrix} v_1\\ v_2\\ \vdots \\ v_n\end{bmatrix} \qquad \vec{v}^T = \begin{bmatrix} v_1 & v_2 & \cdots & v_n\end{bmatrix}$$

## Vector Addition and Subtraction

Addition and subtraction of vectors is done component-wise:

$$\vec{u} + \vec{v} = \begin{bmatrix} u_1 + v_1 \\ u_2 + v_2\\ \vdots \\ u_n + v_n \end{bmatrix} \quad \vec{u} - \vec{v} = \begin{bmatrix} u_1 - v_1 \\ u_2 - v_2\\ \vdots\\ u_n - v_n \end{bmatrix}$$

> **Example:** Consider the vectors $\vec{u}$, $\vec{v}$, and $\vec{w}$ below. For those that are compatible, compute the sums and differences of the vectors. For those that are *not* compatible, describe why.
>

### Comment: Geometric Investigation of Vector Sums and Differences

We often say that vector addition is conducted *head to tail*. This phrase refers to the *geometry* of vector addition. Consider the plot below which shows the sum of the vectors $\vec{u}$ and $\vec{v}$ where $\vec{u} = \begin{bmatrix} 2\\ 1\end{bmatrix}$ and $\vec{v} = \begin{bmatrix} -1\\ 4\end{bmatrix}$.

```python
# @title
import numpy as np
import matplotlib.pyplot as plt

# Vectors
u = np.array([2, 1])
v = np.array([-1, 4])
u_plus_v = u + v

ox = 0
oy = 0

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].quiver(ox, oy, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\vec{u}$')
ax[0].quiver(ox, oy, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\vec{v}$')

ax[0].set_xlim(-5, 5)
ax[0].set_ylim(-1, 7)
ax[0].set_aspect('equal')
ax[0].grid()
ax[0].axhline(0, color='black', linewidth=1.5)
ax[0].axvline(0, color='black', linewidth=1.5)
ax[0].annotate(r'$\vec{u}$', (u[0], u[1]), fontsize = 18)
ax[0].annotate(r'$\vec{v}$', (v[0], v[1]), fontsize = 18)

ax[0].legend(fontsize = 18)
ax[0].set_title(r"Vectors $\vec{u}$ and $\vec{v}$")

ax[1].quiver(ox, oy, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\vec{u}$')
ax[1].quiver(ox, oy, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\vec{v}$')
ax[1].quiver(ox, oy, u_plus_v[0], u_plus_v[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\vec{u} + \vec{v}$')

ax[1].plot([u[0], u_plus_v[0]], [u[1], u_plus_v[1]], 'b--', alpha=0.5)
ax[1].plot([v[0], u_plus_v[0]], [v[1], u_plus_v[1]], 'r--', alpha=0.5)

ax[1].set_xlim(-2, 5)
ax[1].set_ylim(-1, 6)
ax[1].set_aspect('equal')
ax[1].grid()
ax[1].axhline(0, color='black', linewidth=1.5)
ax[1].axvline(0, color='black', linewidth=1.5)

ax[1].annotate(r'$\vec{u}$', (u[0], u[1]), fontsize = 18)
ax[1].annotate(r'$\vec{v}$', (v[0], v[1]), fontsize = 18)
ax[1].annotate(r'$\vec{u} + \vec{v}$', (u_plus_v[0], u_plus_v[1]), fontsize = 18)

ax[1].legend(fontsize = 18)
ax[1].set_title("Vector Addition via Head-to-Tail")

plt.tight_layout()
plt.show()
```

## Scalar Multiplication

Multiplying a vector by a scalar stretches, shrinks, and/or reflects the vector. For example if $a$ is a scalar and $\vec{v}$ is a vector, then

$$a \vec{v} = \begin{bmatrix} a v_1 \\ a v_2\\ \vdots\\ av_n \end{bmatrix}$$

**Example:** Consider the vector $\vec{v} = \begin{bmatrix} 2\\ 1\end{bmatrix}$ and scalars $a_1 = 3$, $a_2 = 0.5$, and $a_3 = -1.5$.

1. $a_1\vec{v} = 3\begin{bmatrix} 2\\ 1\end{bmatrix} = \begin{bmatrix} 3\left(2\right)\\ 3\left(1\right)\end{bmatrix} = \begin{bmatrix} 6\\ 3\end{bmatrix}$
2. $a_2\vec{v} = 0.5\begin{bmatrix} 2\\ 1\end{bmatrix} = \begin{bmatrix} 0.5\left(2\right)\\ 0.5\left(1\right)\end{bmatrix} = \begin{bmatrix} 1\\ 0.5\end{bmatrix}$
3. $a_3\vec{v} = -1.5\begin{bmatrix} 2\\ 1\end{bmatrix} = \begin{bmatrix} -1.5\left(2\right)\\ -1.5\left(1\right)\end{bmatrix} = \begin{bmatrix} -3\\ -1.5\end{bmatrix}$

```python
# @title
v = np.array([2, 1])
scalars = [0.5, 1, 2]
colors = ['gray', 'blue', 'red']

stretch_v = 3*v
shrink_v = 0.5*v
reflect_v = -1.5*v

fig, ax = plt.subplots()

ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale = 1, width = 0.015, color='grey', label=r'$\vec{v}$')
ax.quiver(0, 0, stretch_v[0], stretch_v[1], angles='xy', scale_units='xy', scale=1, color="blue", label=r'$3\vec{v}$')
ax.quiver(0, 0, shrink_v[0], shrink_v[1], angles='xy', scale_units='xy', scale=1, color="red", label=r'$\frac{1}{2}\vec{v}$')
ax.quiver(0, 0, reflect_v[0], reflect_v[1], angles='xy', scale_units='xy', scale=1, color="green", label=r'$-1.5\vec{v}$')

ax.set_xlim(-8, 8)
ax.set_ylim(-8, 8)
ax.set_aspect('equal')
ax.axhline(0, color='black', linewidth=1.5)
ax.axvline(0, color='black', linewidth=1.5)
ax.grid()
ax.legend(loc = "lower right")
ax.set_title("Scalar Multiplication of a Vector")
plt.show()
```

## Vector Multiplication?

We can define multiple types of vector products here.



### Element-Wise Product (Hadamard Product)

We won't make use of this one in linear algebra, but it is an element-wise vector product. It has uses in neural networks, signal processing, optimization and regularization, probablistic modeling, and more.

**Hadamard Product:** Given two vectors $\vec{u}$ and $\vec{v}$ of the same size, we can define the *Hadamard Product* of $\vec{u}$ and $\vec{v}$ to be
$$\vec{u} \circ \vec{v} = \begin{bmatrix} u_1\\ u_2\\ \vdots \\ u_n\end{bmatrix} \circ \begin{bmatrix} v_1\\ v_2\\ \vdots \\ v_n\end{bmatrix} = \begin{bmatrix} u_1v_1\\ u_2v_2\\ \vdots \\ u_nv_n \end{bmatrix}$$

> **Example:** Compute the *Hadamard Product* between the two vectors below.
>
> $$\begin{bmatrix} 1\\ 0\\ -3\\ -5\end{bmatrix} \circ \begin{bmatrix} -3\\ 2\\ -1\\ -3\end{bmatrix}$$

### Inner Product / Dot Product

We'll find many uses for the *dot product* between two vectors. Perhaps you've already encountered some of the uses! For example, the dot product between two vectors $\vec{u}$ and $\vec{v}$ can be used to determine the ***angle*** between the vectors -- in particular, we can use the dot product to quickly determine whether two vectors are *orthogonal* (they meet at a $90^\circ$ angle). The dot product of a vector with itself can also be used to determine the ***length*** of the vector.

**Dot Product:** Given two vectors $\vec{u}$ and $\vec{v}$ of the same size, we can define the *dot product* of $\vec{u}$ and $\vec{v}$ to be

$$\vec{u}\cdot \vec{v} = \begin{bmatrix} u_1\\ u_2\\ \vdots\\ u_n\end{bmatrix}\cdot \begin{bmatrix} v_1\\ v_2\\ \vdots\\ v_n\end{bmatrix} = u_1v_1 + u_2v_2 + \cdots + u_nv_n$$

> 1. Given $\vec{u_1} = \begin{bmatrix} 2\\ -3\end{bmatrix}$ and $\vec{u_2} = \begin{bmatrix} 8\\ -4\end{bmatrix}$, the dot product $\vec{u_1}\cdot\vec{u_2}$ is...
>
>
> 2. $\vec{v_1} = \begin{bmatrix} -9\\ 7\\ 0\\ 1\end{bmatrix}$ and $\vec{v_2} = \begin{bmatrix} 1\\ -1\\ 5\\ -2\end{bmatrix}$, the dot product $\vec{v_1}\cdot \vec{v_2}$ is given by...

**Important Remark:** The *dot product* satisfies the equation

$$\cos\left(\theta\right) = \frac{\vec{u}\cdot\vec{v}}{||\vec{u}||||\vec{v}||}$$

where $\theta$ is the angle between the vectors $\vec{u}$ and $\vec{v}$, and the notation $||\vec{x}||$ denotes the magnitude of the vector $\vec{x}$ and is calculated as $\displaystyle{||\vec{x}|| = \sqrt{\overline{\left(\vec{x}\cdot\vec{x}\right)}}}$.

### Cross Product (3D only)

**Cross Product:** When in three dimensions, we can define the *cross product* between two vectors $\vec{u}$ and $\vec{v}$. The result of taking the cross product is a new vector $\vec{w}$ which is orthogonal (perpendicular) to both $\vec{u}$ and $\vec{v}$.

> ***Note*:** Since there are two opposing directions such a vector could point in, we use the "right-hand rule" to determine which direction the cross-product points.

$$\vec{u} \times \vec{v} = \begin{bmatrix} u_1\\ u_2\\ u_3\end{bmatrix}\times \begin{bmatrix} v_1\\ v_2\\ v_3\end{bmatrix} = \begin{bmatrix} u_2 v_3 - u_3 v_2 \\ u_3 v_1 - u_1 v_3 \\ u_1 v_2 - u_2 v_1 \end{bmatrix}$$

## Vector Norms

**Norms** are generally used to measure the size or magnitude of the object that they are applied to. One *norm* you are familiar with is the absolute value function for real numbers. We can define lots of *norms* for objects, but for vectors in Linear Algebra, we'll be interested in the $\ell_2$ (or Euclidean) norm.

The **$\ell_2$ (Euclidean) norm** of a vector $\vec{v}$ measures its magnitude, and is given by

$$\|\vec{v}\| = \sqrt{\overline{\left(\vec{v}\cdot\vec{v}\right)}} = \sqrt{\overline{\left(v_1^2 + v_2^2 + \cdots + v_n^2\right)}}$$

## Special Vectors

With the vector operations defined above, we can define a few special classes of vector that we'll encounter later in our course (but that you may have already encountered elsewhere).



### Unit Vectors

A vector of magnitude $1$, that is a vector $\vec{v}$ such that $||\vec{v}|| = 1$ is called a **unit vector**.

**Constructing a Unit Vector:** Given any vector $\vec{w}$, which is not the zero-vector, we can construct the *unit vector* pointing in the direction of $\vec{w}$ using

$$\vec{u_w} = \frac{\vec{w}}{||\vec{w}||}$$

> **Example:** Given the vector $\vec{v} = \begin{bmatrix} 2\\ 8\\ -3\end{bmatrix}$, construct a *unit vector* pointing in the same direction as $\vec{v}$.

### Normal Vectors

A vector $\vec{n}$ that is *orthogonal* (perpendicular) to a surface or another vector is called a **normal vector**.

To find vectors normal to a given vector, we remember that $\vec{v}$ and $\vec{n}$ are *orthogonal* if $\vec{v}\cdot \vec{n} = 0$.

> **Examples:** We'll hold off on the construction of *orthogonal*/*normal* vectors until later in our course. We do have some tools which are effective for constructing normal vectors already, and we'll encounter more later on.
>
> + Given a single vector, we can use the fact that the dot product between that vector and a vector normal to it will be $0$.
> + Given a single vector, we can multiply be an appropriate rotation matrix in order to obtain a vector normal to the original.
> + We can use the *cross-product* to construct a vector orthogonal to two vectors in three dimensions.
> + Later in our course, we'll be exposed to the Gram-Schmidt orthogonalization process which will help us in scenarios with larger collections of vectors and/or higher dimensions.
