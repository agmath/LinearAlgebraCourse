# Day 17: Bases and Coordinate Systems

In our Day 10 and 11 notebooks, we introduced the notion of *spans* and *linear independence*, respectively. Given a space and a collection of vectors, we asked and answered questions like:

1. Does the collection span the space?
2. Is the collection linearly independent?

You might intuitively understand these questions as:

1. Does the collection of vectors contain enough information to reconstruct the entire space?
2. Does the collection of vectors contain any redundant information?

It becomes natural now to ask whether we can obtain a collection of vectors with exactly enough information to reconstruct a space without having any redundancies. Such a collection is called a *basis* for the space.

### Bases and Dimension

**Definition (Basis):** A *basis* for any subspace $H$ of $\mathbb{R}^n$ is a linearly independent set of vectors from $H$ which span $H$. Sometimes we say that a basis is a linearly independent spanning set.

**Definition (Dimension):** The *dimension* of a space is the number of vectors in a basis for it.

In the warmup problems you completed today, you noted that the four vectors from $\mathbb{R}^4$ did spanned $\mathbb{R}^4$ but the three vectors from $\mathbb{R}^4$ could not. We can also check that the four vectors from the first wamrup problem are linearly independent, and so the collection is a *basis* for $\mathbb{R}^4$. We've just formalized the notion of dimension -- a descriptor that we've used often, but perhaps hadn't had a rigorous way to define up until now.

**Theorem:** Any linearly independent collection of $n$ vectors from $\mathbb{R}^n$ is a *basis* for $\mathbb{R}^n$.


#### Standard Basis Vectors

The *standard basis* for $\mathbb{R}^n$ is $\mathscr{B} = \left\{\begin{bmatrix} 1\\ 0\\ 0\\ \vdots\\ 0\\ 0\end{bmatrix}, \begin{bmatrix} 0\\ 1\\ 0\\ \vdots\\ 0\\ 0\end{bmatrix}, \cdots, \begin{bmatrix} 0\\ 0\\ 0\\ \vdots\\ 0\\ 1\end{bmatrix}\right\}$. To make this more concrete, consider the standard bases for $\mathbb{R}^2$ and $\mathbb{R}^3$ below.

$$\mathscr{B}_2 = \left\{\begin{bmatrix} 1\\ 0\end{bmatrix}, \begin{bmatrix} 0\\ 1\end{bmatrix}\right\}~~~~~\mathscr{B}_3 = \left\{\begin{bmatrix} 1\\ 0\\ 0\end{bmatrix}, \begin{bmatrix} 0\\ 1\\ 0\end{bmatrix}, \begin{bmatrix} 0\\ 0\\ 1\end{bmatrix}\right\}$$

**Standard Basis Vectors and the Identity Matrix:** Notice that the standard basis vectors for $\mathbb{R}^n$ are the columns of the identity matrix $I_n$.

> ***Note*:** We encountered these vectors previously and, given a space, we've labeled $\vec{e_i}$ as the $i^{\text{th}}$ column of the identity matrix in that space.

The vector notation that we've been utilizing so far in our course depends on these standard basis vectors. For example, we've assumed that the vector $\vec{v} = \begin{bmatrix} 5\\ -4\\ 2\end{bmatrix}$ is the vector $5\vec{e_1} + \left(-4\right)\vec{e_2} + 2\vec{e_3}$. That is, the location of this vector is arrived at by starting at the origin, moving $5$ times in the direction and magnitude of $\vec{e_1}$, $4$ times in the direction and magnitude opposite $\vec{e_2}$, and twice in the direction and magnitude of $\vec{e_3}$.

> **Example:** Consider the vector $\vec{v} = \begin{bmatrix} 7\\ -3\end{bmatrix}$. Identify the space ($\mathbb{R}^n$) that the vector sits in, identify the standard basis vectors for this space, and describe the location of the vector $\vec{v}$ using those standard basis vectors.

## Coordinate Systems

As we mentioned, the importance of having a *basis* is that any vector in the corresponding space has a unique decomposition in terms of the basis vectors. A general spanning set does not provide this uniqueness.

> **Example:** The set $\left\{\left[\begin{array}{r} 1\\ 0\end{array}\right], \left[\begin{array}{r} 1\\ 1\end{array}\right], \left[\begin{array}{r} 0\\ 1\end{array}\right]\right\}$ *spans* $\mathbb{R}^2$.  The vector $\left[\begin{array}{r}5\\ -2\end{array}\right]$ can be written as $5\left[\begin{array}{r}1\\ 0\end{array}\right] - 2\left[\begin{array}{r} 0\\ 1\end{array}\right]$ and also as $5\left[\begin{array}{r} 1\\ 1\end{array}\right] - 7\left[\begin{array}{r} 0\\ 1\end{array}\right]$ (and infinitely many other ways too!). The set $\left\{\left[\begin{array}{r} 1\\ 0\end{array}\right], \left[\begin{array}{r} 0\\ 1\end{array}\right]\right\}$ also *spans* $\mathbb{R}^2$, but since the vectors are linearly independent, the set forms a *basis* for $\mathbb{R}^2$. Using this basis, there is only one way to decompose the vector $\left[\begin{array}{r} 5\\ -2\end{array}\right]$. It is $5\left[\begin{array}{r} 1\\ 0\end{array}\right] - 2\left[\begin{array}{r} 0\\ 1\end{array}\right]$.

**Note:** Any set of two linearly independent vectors from $\mathbb{R}^2$ form a basis for $\mathbb{R}^2$. Some bases are more intuitive and efficient than others, though. Remember our "graph paper" example from our Day 3 notebook. The plot on the left below shows "graph paper" for $\mathbb{R}^2$ corresponding to the standard basis $\mathscr{B}_S = \left\{\left[\begin{array}{r} 1\\ 0\end{array}\right], \left[\begin{array}{r} 0\\ 1\end{array}\right]\right\}$. The middle plot shows "graph paper" for $\mathbb{R}^2$ corresponding to the basis $\mathscr{B}_1 = \left\{\left[\begin{array}{r} 1\\ 0\end{array}\right], \left[\begin{array}{r} 1\\ 1\end{array}\right]\right\}$, and the plot on the right for the basis $\mathscr{B}_3 = \left\{\left[\begin{array}{r} -2\\ - 3\end{array}\right], \left[\begin{array}{r} 1\\ 0.5\end{array}\right]\right\}$.

```python
#@title
e1 = np.array([1, 0])
e2 = np.array([0, 1])
v1 = np.array([1, 0])
v2 = np.array([1, 1])
u1 = np.array([-2, -3])
u2 = np.array([1, 0.5])

plt.figure(figsize = (15, 5))

plt.subplot(1, 3, 1)
plt.grid()
plt.axvline(x = 0, color = "black", alpha = 0.5)
plt.axhline(y = 0, color = "black", alpha = 0.5)
plt.arrow(0, 0, e1[0], e1[1], length_includes_head = True, width = 0.15, head_width = 0.4, color = "C0")
plt.arrow(0, 0, e2[0], e2[1], length_includes_head = True, width = 0.15, head_width = 0.4, color = "C0")

for i in range(-7, 8):
  for j in range(-7, 8):
    curr_loc = np.array([i*e1[0] + j*e2[0], i*e1[1] + j*e2[1]])
    plt.arrow(curr_loc[0], curr_loc[1], e1[0], e1[1], length_includes_head = True, width = 0.1, head_width = 0.1, color = "C0", ls = "--", alpha = 0.3)
    plt.arrow(curr_loc[0], curr_loc[1], e2[0], e2[1], length_includes_head = True, width = 0.1, head_width = 0.1, color = "C0", ls = "--", alpha = 0.2)

plt.xlim((-5, 5))
plt.ylim((-5, 5))
plt.title("B_S Span R2")

plt.subplot(1, 3, 2)
vals = np.linspace(-6, 6)
x1_ax2 = len(vals)*[0]
x2_ax2 = vals
plt.arrow(0, 0, v1[0], v1[1], length_includes_head = True, width = 0.1, head_width = 0.25, color = "C0")
plt.arrow(0, 0, v2[0], v2[1], length_includes_head = True, width = 0.1, head_width = 0.25, color = "C0")
plt.grid()
plt.plot(vals, x1_ax2, color = "black", alpha = 0.5)
plt.plot(vals, x2_ax2, color = "black", alpha = 0.5)

for i in range(-7, 8):
  for j in range(-7, 8):
    curr_loc = np.array([i*v1[0] + j*v2[0], i*v1[1] + j*v2[1]])
    plt.arrow(curr_loc[0], curr_loc[1], v1[0], v1[1], length_includes_head = True, width = 0.1, head_width = 0.1, color = "C0", ls = "--", alpha = 0.3)
    plt.arrow(curr_loc[0], curr_loc[1], v2[0], v2[1], length_includes_head = True, width = 0.1, head_width = 0.1, color = "C0", ls = "--", alpha = 0.2)

plt.xlim((-4, 4))
plt.ylim((-4, 4))
plt.title("B_1 Spans R2")

plt.subplot(1, 3, 3)
vals = np.linspace(-6, 6)
x1_ax3 = (3/2)*vals
x2_ax3 = 0.5*vals

plt.grid()
plt.plot(vals, x1_ax3, color = "black", alpha = 0.5)
plt.plot(vals, x2_ax3, color = "black", alpha = 0.5)
plt.arrow(0, 0, u1[0], u1[1], length_includes_head = True, width = 0.2, head_width = 0.4, color = "C0")
plt.arrow(0, 0, u2[0], u2[1], length_includes_head = True, width = 0.2, head_width = 0.4, color = "C0")

for i in range(-20, 20):
  for j in range(-20, 20):
    curr_loc = np.array([i*u1[0] + j*u2[0], i*u1[1] + j*u2[1]])
    plt.arrow(curr_loc[0], curr_loc[1], u1[0], u1[1], length_includes_head = True, width = 0.1, head_width = 0.1, color = "C0", ls = "--", alpha = 0.3)
    plt.arrow(curr_loc[0], curr_loc[1], u2[0], u2[1], length_includes_head = True, width = 0.1, head_width = 0.1, color = "C0", ls = "--", alpha = 0.2)

plt.xlim((-6, 6))
plt.ylim((-6, 6))
plt.title("B_2 Spans R2")

plt.show()
```

### Change of Basis

In the fourth warmup problem from today, you provided directions to get from the *House* to the *Store*. Its unlikely that those directions sounded something like "*go about a half mile east and then a mile and a quarter north*". You probably mentioned that the persion should "*travel four blocks north-east on Maple and then turn onto Main and go three blocks north-west*".

In this case, it was much more useful to use the streets than it was to use the usual N-S/E-W cardinal directions. Since it was easier and more effective to navigate in the directions of the roadways, you chose to do that, which is a completely valid and justifiable choice to make.

You performed a *change of basis*.

**Remark:** We can see from the above examples that the choice of basis is not unique. However, once we've chosen a basis, each vector in our space has a unique representation under the basis. Additionally, we see that the choice of basis vectors determines the structure of our space. For example, using bases $\mathscr{B}_2$ or $\mathscr{B}_3$ results in a coordinate system where the axes do not meet at a $90^\circ$ angle. Additionally, using basis $\mathscr{B}_3$ results in a coordinate system where moving *downwards* is growing in the $x_2$ direction. That is, vectors "above" that $x_1$ axis will have a negative second component!

**Definition (Basis Representation of Vectors):** Suppose that $\mathscr{B} = \left\{\vec{b_1}, \vec{b_2}, \cdots, \vec{b_p}\right\}$ is a basis for a subspace $H$. For each $\vec{x}\in H$, the *coordinates of $\vec{x}$ relative to the basis $\mathscr{B}$* are the weights $c_1, c_2, \cdots, c_p$ sucht that $\vec{x} = c_1\vec{b_1} + c_2\vec{b_2} + \cdots + c_p\vec{b_p}$. The vector $\left[\vec{x}\right]_{\mathscr{B}} = \left[\begin{array}{r} c_1\\ c_2\\ \vdots\\ c_p\end{array}\right]$ in $\mathbb{R}^p$ is called the *coordinate vector of $\vec{x}$ relative to $\mathscr{B}$*.

**Completed Example 1:** Consider the basis $\mathscr{B} = \left\{\left[\begin{array}{r} -2\\ 3\end{array}\right], \left[\begin{array}{r} 1\\ 5\end{array}\right]\right\}$ and determine the position of the vector $\left[\vec{x}\right]_{\mathscr{B}} = \left[\begin{array}{r} 5\\ -1\end{array}\right]$ in $\mathbb{R}^2$ under the standard basis $\mathscr{B}_S = \left\{\left[\begin{array}{r} 1\\ 0\end{array}\right], \left[\begin{array}{r} 0\\ 1\end{array}\right]\right\}$.

> *Solution.* The vector $\left[\vec{x}\right]_{\mathscr{B}} = \left[\begin{array}{r} 5\\ -1\end{array}\right]$ is the vector $5\left[\begin{array}{r} -2\\ 3\end{array}\right] + \left(-1\right)\left[\begin{array}{r} 1\\ 5\end{array}\right]$. We can simplify the linear combination to obtain the representation of $\vec{x}$ in the standard basis. That is, $\vec{x} = \left[\begin{array}{r} -11\\ 10\end{array}\right]$ in the standard basis $\mathscr{B}_{S} = \left\{\left[\begin{array}{r} 1\\ 0\end{array}\right], \left[\begin{array}{r} 0\\ 1\end{array}\right]\right\}$. Notice that the standard basis forms our usual "$x$" and "$y$" axes. $_\blacktriangledown$

**Note:** Another approach to converting a vector $\vec{x}$, written in a non-standard basis $\mathscr{B}$, into its equivalent using the standard basis is to construct the matrix $P_{\mathscr{B}}$ whose columns are the vectors in $\mathscr{B}$ and then to note that $\vec{x} = P_{\mathscr{B}}\left[\vec{x}\right]_{\mathscr{B}}$.

**Completed Example 2:** Consider the basis $\mathscr{B} = \left\{\left[\begin{array}{r} -2\\ 3\end{array}\right], \left[\begin{array}{r} 1\\ 5\end{array}\right]\right\}$ and determine the $\mathscr{B}$-coordinates of the vector $\vec{x} = \left[\begin{array}{r} -5\\ 14\end{array}\right]$.

> *Solution.* In order to find the $\mathscr{B}$-coordinates for our vector $\vec{x}$, we need to write $\vec{x}$ as a linear combination of the vectors in $\mathscr{B}$. The $\mathscr{B}$-coordinates of the vector $\vec{x}$ will be the weights from that linear combination. That is, $\left[\vec{x}\right]_{\mathscr{B}} = \left[\begin{array}{r} c_1\\ c_2\end{array}\right]$, where $\vec{x} = c_1\left[\begin{array}{r} -2\\ 3\end{array}\right] + c_2\left[\begin{array}{r} 1\\ 5\end{array}\right]$.
>
> We'll find $c_1$ and $c_2$ by setting up an appropriate augmented matrix, $\left[\begin{array}{rr|r} -2 & 1 & -5\\ 3 & 5 & 14\end{array}\right]$ and row-reducing to solve.
>
> \begin{align} \left[\begin{array}{rr|r} -2 & 1 & -5\\ 3 & 5 & 14\end{array}\right] &\substack{R_{2} \leftarrow 2R_2 + 3R_3\\ \longrightarrow} \left[\begin{array}{rr|r} -2 & 1 & -5\\ 0 & 13 & 13\end{array}\right]\\
&\substack{R_{2} \leftarrow \left(\frac{1}{13}\right)R_2\\ \longrightarrow} \left[\begin{array}{rr|r} -2 & 1 & -5\\ 0 & 1 & 1\end{array}\right]\\
&\substack{R_{1} \leftarrow R_1 + \left(-1\right)R_2\\ \longrightarrow} \left[\begin{array}{rr|r} -2 & 0 & -6\\ 0 & 1 & 1\end{array}\right]\\
&\substack{R_{1} \leftarrow \left(\frac{-1}{2}\right)R_{1}\\ \longrightarrow} \left[\begin{array}{rr|r} 1 & 0 & 3\\ 0 & 1 & 1\end{array}\right]
\end{align}
>
> From here, we see that $c_1 = 3$ and $c_2 = 1$. That is, the $\mathscr{B}$-coordinates for $\vec{x}$ are $\left[\vec{x}\right]_{\mathscr{B}} = \left[\begin{array}{r} 3\\ 1\end{array}\right]$. $_\blacktriangledown$

**Note:** Another approach to rewriting a vector $\vec{x}$ in an alternative basis $\mathscr{B}$ is to construct the matrix $P$ whose columns are the basis vectors, find its inverse, and then note that $\left[\vec{x}\right]_\mathscr{B} = P_{\mathscr{B}}^{-1}\vec{x}$
