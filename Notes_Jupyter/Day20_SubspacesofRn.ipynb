# Day 20: Subspaces of $\mathbb{R}^n$

Recall that $\mathbb{R}^n = \left\{\left[\begin{array}{r} x_1\\ x_2\\ \vdots\\ x_n\end{array}\right] \mid x_i\in\mathbb{R}\right\}$. That is, $\mathbb{R}^n$ is the space consisting of all $n$-component vectors whose entries are real numbers. In this notebook we'll be focused on *subspaces* of $\mathbb{R}^n$ connected with the matrix $A$ and providing useful insight into the matrix equation $A\vec{x} = \vec{b}$.

**Definition (Subspace Criteria):** A *subspace* of $\mathbb{R}^n$ is any set $H$ in $\mathbb{R}^n$ satisfying all of the following properties.

1. $\vec{0} \in H$
2. If $\vec{u}, \vec{v}\in H$, then $\vec{u} + \vec{v}\in H$
3. If $\vec{v}\in H$, then $c\vec{v}\in H$ for any scalar $c$


**Note:** Items (2.) and (3.) in the definition above are often described as the subspace being *closed under addition and scalar multiplication*.

**Completed Example 1:** Show that $\text{span}\left(\left\{\left[\begin{array}{r} 1\\ -3\end{array}\right]\right\}\right)$ is a subspace of $\mathbb{R}^2$.

> *Solution.* In order to show that $\text{span}\left(\left\{\left[\begin{array}{r} 1\\ -3\end{array}\right]\right\}\right)$ is a subspace of $\mathbb{R}^2$, we'll need to show the three subspace properties above.
>
> + Note that $\vec{0}\in \text{span}\left(\left\{\left[\begin{array}{r} 1\\ -3\end{array}\right]\right\}\right)$ since $\vec{0} = 0\cdot\left[\begin{array}{r} 1\\ -3\end{array}\right]$. $\checkmark$
> + Let $\vec{x}$ and $\vec{y}$ be in $\text{span}\left(\left\{\left[\begin{array}{r} 1\\ -3\end{array}\right]\right\}\right)$. Then, there exist scalars $c_1$ and $c_2$ such that $\vec{x} = c_1\left[\begin{array}{r} 1\\ -3\end{array}\right]$ and $\vec{y} = c_2\left[\begin{array}{r} 1\\ -3\end{array}\right]$. Now, notice that
>
> \begin{align*} \vec{x} + \vec{y} &= c_1\left[\begin{array}{r} 1\\ -3\end{array}\right] + c_2\left[\begin{array}{r} 1\\ -3\end{array}\right]\\
&= \left(c_1 + c_2\right)\left[\begin{array}{r} 1\\ -3\end{array}\right]
\end{align*}
>
> > Note that $c_1 + c_2$ is another scalar $c_3$. Thus, $\vec{x} + \vec{y}$ is in $\text{span}\left(\left\{\left[\begin{array}{r} 1\\ -3\end{array}\right]\right\}\right)$. $\checkmark$
>
> + Finally, let $c$ be any scalar and let $\vec{x}$ be in $\text{span}\left(\left\{\left[\begin{array}{r} 1\\ -3\end{array}\right]\right\}\right)$. Then, there exists another scalar $d$ such that $\vec{x} = d\left[\begin{array}{r} 1\\ -3\end{array}\right]$. Now,
> \begin{align*} c\vec{x} &= c\left(d\left[\begin{array}{r} 1\\ -3\end{array}\right]\right)\\
&= \left(cd\right)\left[\begin{array}{r} 1\\ -3\end{array}\right]
\end{align*}
>> Note that $cd$ is a scalar, so $c\vec{x}$ is in $\text{span}\left(\left\{\left[\begin{array}{r} 1\\ -3\end{array}\right]\right\}\right)$. $\checkmark$
>
> Since the three subspace properties hold, we have that $\text{span}\left(\left\{\left[\begin{array}{r} 1\\ -3\end{array}\right]\right\}\right)$ is a subspace of $\mathbb{R}^2$. $_\blacksquare$

**Definition (Column Space):** The *column space* corresponding to the matrix $A$, denoted by $\text{col}\left(A\right)$ is the set of all linear combinations of the columns of $A$. That is, $\text{col}\left(A\right) = \text{span}\left(\left\{\vec{a_1}, \vec{a_2}, \cdots, \vec{a_n}\right\}\right)$ where $\vec{a_i}$ is the $i^{th}$ column of the matrix $A$.

**Theorem (Column Space is a Subspace):** If $A$ is an $m\times n$ matrix, then $\text{col}\left(A\right)$ is a subspace of $\mathbb{R}^m$.

**Definition (Null Space):** The *null space* of a matrix $A$ is denotel by $\text{Nul}\left(A\right)$ and is the set of all solutions to the homogeneous equation $A\vec{x} = \vec{0}$.

**Theorem (Null Space is a Subspace):** For any $m\times n$ matrix $A$, $\text{Nul}\left(A\right)$ is a subspace of $\mathbb{R}^n$.

**Theorem (Basis for $\text{col}\left(A\right)$):** The pivot columns of $A$ form a basis for $\text{col}\left(A\right)$.

> **Note (Important!):** In the Theorem listed above, it is important that we use pivot columns of the original matrix $A$ as a basis for its column space. Often times the pivot columns of a row reduced version of $A$ will not span $\text{col}\left(A\right)$.

**Completed Example 2:** Find a basis for the null space of the matrix $A = \left[\begin{array}{rrr} 1 & 2 & 0\\ 0 & 1 & 1\end{array}\right]$.

> *Solution.* We'll start by solving the homogeneous equation $A\vec{x} = \vec{0}$.
> \begin{align*} \left[\begin{array}{rrr|r} 1 & 2 & 0 & 0\\ 0 & 1 & 1 & 0\end{array}\right] \substack{R_1\leftarrow R_1 + (-2R2)\\ \longrightarrow} \left[\begin{array}{rrr|r} 1 & 0 & -2 & 0\\ 0 & 1 & 1 & 0\end{array}\right]
\end{align*}
> From the reduced-row-echelon form matrix above, we can identify that $x_3$ is a *free variable*. The solution vector is then $\vec{x} = \left[\begin{array}{r} 2x_3\\ -x_3\\ x_3\end{array}\right]$, which we can write in parametric vector form as $\vec{x} = x_3\left[\begin{array}{r} 2\\ -1\\ 1\end{array}\right]$. This means that any scalar multiple of $\left[\begin{array}{r} 2\\ -1\\ 1\end{array}\right]$ is in $\text{Nul}\left(A\right)$. That is a basis for $\text{Nul}\left(A\right)$ is $\mathscr{B}_{\text{Nul}\left(A\right)} = \left\{\left[\begin{array}{r} 2\\ -1\\ 1\end{array}\right]\right\}$. $_\blacktriangledown$

**Completed Example 3:** Find a basis for the column space of the matrix $A = \left[\begin{array}{rrr} 1 & 3 & -2\\ -1 & 1 & 5\\ 0 & 4 & 3\end{array}\right]$.

> *Solution.* We'll start by identifying the pivot columns in the matrix $A$.
> \begin{align*} \left[\begin{array}{rrr} 1 & 3 & -2\\ -1 & 1 & 5\\ 0 & 4 & 3\end{array}\right] \substack{R_2 \leftarrow R_2 + R_1\\ \longrightarrow} \left[\begin{array}{rrr} 1 & 3 & -2\\ 0 & 4 & 3\\ 0 & 4 & 3\end{array}\right]\\
\substack{R_3 \leftarrow R_3 + (-1R_2)\\ \longrightarrow} \left[\begin{array}{rrr} 1 & 3 & -2\\ 0 & 4 & 3\\ 0 & 0 & 0\end{array}\right]\\
\end{align*}
>
> Note that the first and second columns are pivot columns in the row-echelon form of the matrix. This means that the first and second columns of the original matrix $A$ are pivot columns, and so these columns form a basis for $\text{Col}\left(A\right)$. That is, $\mathscr{B}_{\text{Col}\left(A\right)} = \left\{\left[\begin{array}{r} 1\\ -1\\ 0\end{array}\right], \left[\begin{array}{r} 3\\ 1\\ 4\end{array}\right]\right\}$. $_\blacktriangledown$

### Dimension, Rank, Null Space, and Column Space

We'll close by highlighting a few important observations about the ideas and objects we've introduced in this notebook.

+ The *rank* of a matrix $A$ is the number of pivot positions in $A$.
+ $\dim\left(\text{Col}\left(A\right)\right) = \text{rank}\left(A\right)$
+ Given an $n\times n$ matrix $A$, we have $\dim\left(\text{Col}\left(A\right)\right) + \dim\left(\text{Null}\left(A\right)\right) = n$

## Applications

**Definition (Vector Space):** A *vector space* is a non-empty set $V$ of objects, called *vectors*, on which are defined two operations, called *addition* and *scalar multiplication*, where the ten properties below hold for all $\vec{u},~\vec{v},~\vec{w}\in V$ and any scalars $c$ and $d$.

1. $\vec{u} + \vec{v} \in V$
2. $\vec{u} + \vec{v} = \vec{v} + \vec{u}$
3. $\vec{u} + \left(\vec{v} + \vec{w}\right) = \left(\vec{u} + \vec{v}\right) + \vec{w}$
4. There is a $\vec{0}\in V$ such that $\vec{v} + \vec{0} = \vec{0} + \vec{v} = \vec{v}$.
5. For each $\vec{u}\in V$ there is a $-\vec{u} \in V$ such that $\vec{u} + \left(-\vec{u}\right) = \vec{0}$
6. $c\vec{u} \in V$
7. $c\left(\vec{u} + \vec{v}\right) = c\vec{u} = c\vec{v}$
8. $\left(c + d\right)\vec{u} = c\vec{u} + d\vec{u}$
9. $c\left(d\vec{u}\right) = \left(cd\right)\vec{u}$
10. $1\vec{u} = \vec{u}$
